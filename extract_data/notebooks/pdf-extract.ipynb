## Extracting Text

```python
from pypdf import PdfReader

reader = PdfReader("/Users/dhruvsharma/Projects/catalog-to-pdf/baldwing-catalog.pdf")
print(len(reader.pages))

page_3 = reader.pages[3]

print(page_3.extract_text())

with open("PyPDF.txt", "w") as file:
    file.write(page_3.extract_text())

for i in page_3.images:
    with open(i.name, "wb") as file:
        file.write(i.data)
```

## Testing with OCR

```python
import os

# Ensure TESSDATA_PREFIX is set correctly
os.environ['TESSDATA_PREFIX'] = '/opt/anaconda3/envs/catalog/share/tessdata/'

# Verify that TESSDATA_PREFIX is set
print(f"TESSDATA_PREFIX: {os.getenv('TESSDATA_PREFIX')}")

try:
    doc = pymupdf.open("baldwing-catalog.pdf")
    page = doc.load_page(7) 

    full_tp = page.get_textpage_ocr(flags=0, dpi=300, full=False)
    print(page.get_text(textpage=full_tp))
    
    with open("PyMuPDFOCR.txt", "w") as file:
        file.write(page.get_text(textpage=full_tp))

except RuntimeError as e:
    print(f"RuntimeError: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```
